##Importando as bibliotecas

```python
import pandas as pd
import yfinance as yf
```

Obtenha primeiramente o arquivo da B3 antes de executar o passo a seguir

```python
composicao_ibov = pd.read_csv('/content/IBOVDia_24-08-23.csv', encoding = 'latin-1', header = 1, sep = ';')
```

```python
composicao_ibov
```

```python
composicao_ibov.index[0:-2] # retira as duas ultimaslinhas
```

```python
tickers = list(composicao_ibov.index[0:-2])
```

```python
tickers_novo = [acao + '.SA' for acao in tickers] # compressão de lista
```

```python
ibov = yf.download(tickers_novo, start = '2023-01-01')['Close']
```

```python
ibov.plot(figsize=(15,8));
```

```python
ibov.to_csv('cotacoes_ibov.csv')
```
